```python
# Assignment: Job Applicant Screening System
# A company wants to develop a screening system for job applicants based on the following criteria:
# Inputs:
# ● Age (integer)
# ● Years of Experience (integer)
# ● Highest Qualification (string: one of "HighSchool", "Bachelors", "Masters", "PhD")
# ● Certifications (string: one of "None", "Some", "Many")
# ● Background Check Passed? (string: "Yes" or "No")
# Screening Decision Rules:
# 1. Age restriction:
# ○ If the applicant is under 18 years old, reject the application.
# ■ Message: "Rejected: Must be at least 18 years old."
# 2. Senior age experience check:
# ○ If the applicant is over 60 years old and has less than 10 years of experience, reject the application.
# ■ Message: "Rejected: Insufficient experience for senior age."
# 3. Background check:
# ○ If the applicant fails the background check, reject the application.
# ■ Message: "Rejected: Background check failed."
# 4. Qualification and experience-based screening:
# ○ If the applicant's qualification is "PhD" or "Masters":
# ■ If experience is 5 years or more, approve the application.
# ■ Message: "Approved: Senior level candidate."
# ■ Else if experience is between 2 and 5 years (inclusive lower bound,
# exclusive upper bound):
# ■ If certifications are "Many", approve the application.
# ■ Message: "Approved: Promising candidate with
# certifications."
# ■ Else, reject the application.
# ■ Message: "Rejected: Insufficient experience
# and certifications."
# ■ Else, reject the application.
# ■ Message: "Rejected: Experience too low for senior
# qualification."
# ○ Else if the qualification is "Bachelors":
# ■ If experience is 3 years or more, approve the application.
# ■ Message: "Approved: Qualified candidate."
# ■ Else if certifications are "Many", approve the application.
# ■ Message: "Approved: Candidate with
# certifications."
# ■ Else, reject the application.
# ■ Message: "Rejected: Not enough experience or
# certifications."
# ○ Else if the qualification is "HighSchool":
# ■ If experience is 5 years or more and certifications are not "None",
# approve the application.
# ■ Message: "Approved: Experienced candidate with
# certifications."
# ■ Else, reject the application.
# ■ Message: "Rejected: Does not meet minimum
# qualification."
# This system uses multiple levels of nested conditions to carefully evaluate each applicant based
# on their age, experience, qualification, certifications, and background check status.
# Test case:
# Input: Age=50, Experience=6, Qualification='HighSchool', Certifications='Some',
# Background='Yes' Expected Output: Approved: Experienced candidate with certifications
# Input: Age=40, Experience=4, Qualification='HighSchool', Certifications='None',
# Background='Yes' Expected Output: Rejected: Does not meet minimum qualification.
# Input: Age=25, Experience=1, Qualification='PhD', Certifications='Many', Background='Yes'
# Expected Output: Approved: Promising candidate with certifications.
# Input: Age=35, Experience=6, Qualification='Masters', Certifications='Some', Background='Yes'
# Expected Output: Approved: Senior level candidate.
```

```python
#### Job Applicant Screening System

# Input
age = int(input("Enter Age: "))
experience = int(input("Enter Years of Experience: "))
qualification = input("Enter Highest Qualification (HighSchool/Bachelors/Masters/PhD): ")
certifications = input("Enter Certifications (None/Some/Many): ")
background = input("Background Check Passed? (Yes/No): ")

# Screening Logic

# 1. Age restriction
if age < 18:
    print("Rejected: Must be at least 18 years old.")

# 2. Senior age experience check
elif age > 60 and experience < 10:
    print("Rejected: Insufficient experience for senior age.")

# 3. Background check
elif background == "No":
    print("Rejected: Background check failed.")

# 4. Qualification and experience-based screening
else:
    if qualification in ["PhD", "Masters"]:
        if experience >= 5:
            print("Approved: Senior level candidate.")
        elif experience >= 2 and experience < 5:
            if certifications == "Many":
                print("Approved: Promising candidate with certifications.")
            else:
                print("Rejected: Insufficient experience and certifications.")
        else:
            print("Rejected: Experience too low for senior qualification.")

    elif qualification == "Bachelors":
        if experience >= 3:
            print("Approved: Qualified candidate.")
        elif certifications == "Many":
            print("Approved: Candidate with certifications.")
        else:
            print("Rejected: Not enough experience or certifications.")

    elif qualification == "HighSchool":
        if experience >= 5 and certifications != "None":
            print("Approved: Experienced candidate with certifications.")
        else:
            print("Rejected: Does not meet minimum qualification.")

    else:
        print("Rejected: Invalid qualification entered.")
```
